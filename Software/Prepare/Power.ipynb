```python
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
plt.rcParams.update({
    'text.usetex' : True
})
```

```python
k_B =  1.380649e-23 / 8.6173E-11 #J/MeV
E_fus = 17.6 #MeV
aplph = 7.2973525693e-3
r_e = 2.81794092e-15
m_e = 9.10938215e-31
c = 3*10^8
Z_eff = 1
```

```python
T = 1e-6
n_e = 2e18
n_T = 1e18

```

```python
def N(P_e, Tau):
    return 2*P_e/ (3*k_B*T/Tau + 8.5*aplph*r_e**2*m_e*c**3*Z_eff**2*n_e*np.sqrt(k_B*T/(m_e*c**2)) - 0.1*n_T*1.1*10e-24*T**2*E_fus)
```

```python
M_Tau = np.linspace(1e-4,1e-2, 10**(7))
N_r = N(3000,M_Tau)
```

```python
plt.plot(M_Tau,N_r, color="r")
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Время удержания энергии, с")
plt.ylabel("Количество ионов")
#plt.title("Number of Ions by Time")
plt.show()
```

```python
V = (N_r/2)/n_T
V = V/1e-6
plt.plot(M_Tau,V, color="r", label="n = 1e9 см^-3")
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Время удержания плазмы, с")
plt.ylabel("Объем, мм^3")
# plt.title("Volume of generator by External Power")
plt.show()
```

```python
def V_ion(T, M_X):
    E = 3*k_B*T
    return np.sqrt(c**2*(1 - M_X**2*c**4/(E+M_X*c**2)**2))
```

```python
M_P = 1.6726e-27
M_N = 1.6749e-27
M_D = M_P + M_N 
M_T = M_P + 2*M_N
```

```python
V_D = V_ion(T, M_D)
V_T = V_ion(T,M_T)
```

```python
print(V_D/c)
```

```python
F_D = N_r*M_D*V_D/(M_Tau*np.sqrt(1-(V_D*V_D)/c**2))
F_T = N_r*M_T*V_T/(M_Tau*np.sqrt(1-(V_T*V_T)/c**2))
F_r = F_D + F_T
```

```python
plt.plot(M_Tau, F_r, color="r")
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Holding time, s")
plt.ylabel("Force, N")
# plt.title("Force by Time")
plt.show()
```

```python
print(max(F_r))
print(min(F_r))
```

```python
consumption = N_r/M_Tau*(M_D + M_T)
```

```python

plt.plot(M_Tau, consumption, color="r", label="T = 1e-6 MeV")
plt.xlabel("Holding Time, s")
plt.ylabel("consumption, kg")
plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.show()
```

```python
consToForce = consumption/F_r
```

```python

plt.plot(M_Tau, consToForce, color="r", label="T = 1e-6 MeV")
plt.xlabel("Holding Time, s")
plt.ylabel("Fuel consumption per unit of force, kg/N")
#plt.xscale("log")
#plt.yscale("log")
plt.legend()
plt.show()
```

```python
Ship_mass = 1e7
```

```python
def Cial_Vel(V,M,M_0):
    return V*np.log(M/M_0)
```

```python
def DeltaV(N_DS, N_TS, V_DS, V_TS):
    return Cial_Vel(V_DS, M_D*N_DS + M_T*N_TS + Ship_mass, Ship_mass + M_T*N_TS) + Cial_Vel(V_TS, M_T*N_TS + Ship_mass, Ship_mass )
```

```python
dV = DeltaV(N_r, N_r, V_D, V_T)
```

```python
plt.plot(M_Tau, dV, color="r", label="T=1e-6 MeV \n Ship Mass = 1e6")
plt.xlabel("Hold Time, s")
plt.ylabel("$\Delta V$")
plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.show()
```

```python
consToDV = consumption/dV
```

```python
c2 = consumption

plt.plot(dV, c2, color="r", label="T=1e-6 MeV \nShip Mass = 1e7")
plt.ylabel("consumption, kg")
plt.xlabel("$\Delta v$, m/s")
#plt.title("Consumption per $\Delta v$")
plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.show()
```

```python
plt.plot(M_Tau, consToDV, color="r", label="T=1e-6 MeV \nShip Mass = 1e7")
plt.xlabel("Hold Time, s")
plt.ylabel("Consumption per $\Delta V$")
plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.show()
```
